```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
sns.set_context("poster")
sns.set(rc={"figure.figsize": (12.,6.)})
sns.set_style("whitegrid")
```

```python
%matplotlib inline
%config Inlinebackend.figure_format= 'retina'
```

```python
df1 = pd.read_csv('data_limpio.csv', encoding='utf-8')
df1.sample(20)
```

## ATAQUES DE TIBURÓN A LO LARGO DE LA HISTORIA

```python
sns.set(rc={"figure.figsize": (18,10)})
sns.countplot(x=df1.año[df1.año != 0]).set(title="Ataques de tiburon en la historia",ylabel="Ataques")
```

## FREQUENCIA DE ATAQUES SEGÚN EL MES DEL AÑO

```python
sns.countplot(x=df1.mes[df1.mes != 0]).set(title="Ataques de tiburon según el mes",ylabel="Ataques")
```

## EDAD DE LAS VICTIMAS

```python
sns.histplot(x=df1.age).set(title="Edad de las victimas",ylabel="Ataques")
```

## PARTES CORPORALES HERIDAS

```python
ax = sns.countplot(x=df1.lugar_lesion)
ax.set(title="Partes dañadas", 
       ylabel="Ataques")
```

## FREQUENCIA DE TIPOS DE LESIONES

```python
ax = sns.countplot(x=df1.tipo_lesion)
ax.set(title="tipo lesiones", 
       ylabel="Ataques")
```

## MORTALIDAD DE LOS ATAQUES

```python
ax = sns.countplot(x=df1["fatality"])
ax.set(title="Fatalidad", 
       ylabel="Ataques")
```

## ATAQUES SEGÚN ESPECIE DE TIBURÓN

```python
ax = sns.countplot(x=df1.shark_specimen)
ax.set(title="Ataques según especie de tiburón", 
       ylabel="Ataques")
```

## MORTALIDAD SEGÚN ESPECIE DE TIBURÓN

```python
fig = df1.groupby(["shark_specimen", "fatality"]).agg({"fatality": "count"}).unstack().plot(kind="bar")
```
