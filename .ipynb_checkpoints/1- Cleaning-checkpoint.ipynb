Dentro de injuries, analizando las palabras que aparecen con frecuencia mayor de 50, puedo seleccionar

a) tipo de lesion

b) parte del cuerpo afectada

hago dos diccionarios para poder pasarlos por la columna y crear dos columnas nuevas con valores categoricos.


```python
tipo_lesiones ={'bitten'      : 'mordiscos'  ,
                'bit'         : 'mordiscos'  ,
                'bite'        : 'mordiscos'  ,
                'lacerations' : 'laceraciÃ³n' ,
                'laceration'  : 'laceraciÃ³n' , 
                'lacerated'   : 'laceraciÃ³n' ,
                'bumped'      : 'golpe'      ,
                'rammed'      : 'golpe'      ,
                'wounds'      : 'herida'     ,
                'amputated'   : 'amputacion' }

lugar_lesion = {'leg'     : 'piernas' ,
                'legs'    : 'piernas' ,
                'thigh'   : 'piernas' ,
                'calf'    : 'piernas' ,
                'knee'    : 'piernas' ,
                'foot'    : 'pies'    ,
                'feet'    : 'pies'    ,
                'heel'    : 'pies'    ,
                'ankle'   : 'pies'    ,
                'body'    : 'tronco'  ,
                'shoulder': 'tronco'  ,
                'torso'   : 'tronco'  ,
                'abdomen' : 'tronco'  ,
                'back'    : 'tronco'  ,
                'chest'   : 'tronco'  ,
                'hip'     : 'tronco'  ,
                'head'    : 'cabeza'  ,
                'arm'     : 'brazos'  ,
                'arms'    : 'brazos'  ,
                'elbow'   : 'brazos'  ,
                'forearm' : 'brazos'  ,
                'hand'    : 'manos'   ,
                'fingers' : 'manos'   ,
                'finger'  : 'manos'   ,
                'hands'   : 'manos'   ,
                'wrist'   : 'manos'   }
```

```python
df1.injury.fillna("others/unknown" , inplace=True)
```

```python
def definir_tipo_lesion(injury) :
    for lesion , clase_lesion in tipo_lesiones.items() :
        if lesion in injury.lower() :
            return clase_lesion
    return 'others/unknown'
```

```python
def definir_lugar_lesion(injury) :
    for lugar, parte_cuerpo in lugar_lesion.items() :
        if lugar in injury.lower() :
            return parte_cuerpo
    return 'others/unknown'
```

```python
df1 = df1.assign(tipo_lesion = df1.injury.apply(definir_tipo_lesion))
```

```python
df1 = df1.assign(lugar_lesion = df1.injury.apply(definir_lugar_lesion));
```
